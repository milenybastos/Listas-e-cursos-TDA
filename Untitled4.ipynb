```python
import random
import os

def limpar_tela():
    # Limpa a tela do terminal
    os.system('cls' if os.name == 'nt' else 'clear')

def exibir_forca(erros):
    estagios = [
        """
           -----
           |   |
               |
               |
               |
               |
        =========
        """,
        """
           -----
           |   |
           O   |
               |
               |
               |
        =========
        """,
        """
           -----
           |   |
           O   |
           |   |
               |
               |
        =========
        """,
        """
           -----
           |   |
           O   |
          /|   |
               |
               |
        =========
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
               |
               |
        =========
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          /    |
               |
        =========
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          / \\  |
               |
        =========
        """,
    ]
    print(estagios[erros])

def jogar_forca():
    limpar_tela()
    print("üéÉ Bem-vindo √† Forca Assustadora! üï∏Ô∏è\n")

    palavras_halloween = [
        "fantasma", "bruxa", "abobora", "vampiro",
        "caveira", "zumbi", "mumia", "morcego",
        "assustador", "caldeirao", "feitico"
    ]

    palavra_secreta = random.choice(palavras_halloween)
    letras_certas = []
    letras_erradas = []
    max_erros = 6
    erros = 0

    while erros < max_erros:
        limpar_tela()
        exibir_forca(erros)

        # Monta a palavra atual
        palavra_atual = ""
        for letra in palavra_secreta:
            if letra in letras_certas:
                palavra_atual += letra + " "
            else:
                palavra_atual += "_ "
        print("\nPalavra:", palavra_atual)
        print("Letras erradas:", " ".join(letras_erradas))

        # Vit√≥ria
        if "_" not in palavra_atual:
            print("\nüïØÔ∏è Parab√©ns! Voc√™ descobriu a palavra secreta! üßô‚Äç‚ôÄÔ∏è")
            break

        tentativa = input("\nDigite uma letra: ").lower()

        if not tentativa.isalpha() or len(tentativa) != 1:
            print("Digite apenas uma letra v√°lida!")
            input("Pressione Enter para continuar...")
            continue

        if tentativa in letras_certas or tentativa in letras_erradas:
            print("Voc√™ j√° tentou essa letra!")
            input("Pressione Enter para continuar...")
            continue

        if tentativa in palavra_secreta:
            letras_certas.append(tentativa)
        else:
            letras_erradas.append(tentativa)
            erros += 1

    # Derrota
    if erros == max_erros:
        limpar_tela()
        exibir_forca(erros)
        print("\nüíÄ Voc√™ foi enfeiti√ßado! A palavra era:", palavra_secreta)

if __name__ == "__main__":
    jogar_forca()
```
