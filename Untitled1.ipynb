```python

import random

def jogar_adivinhacao():
    """
    üéÉ Jogo de Adivinha√ß√£o com tema de Halloween
    A bruxa guarda um n√∫mero secreto em seu caldeir√£o.
    Descubra o n√∫mero antes que o feiti√ßo se complete!
    """

    # 1. Apresenta√ß√£o
    print("üßô‚Äç‚ôÄÔ∏è Bem-vindos √† Cabana da Bruxa!")
    print("A Bruxa de Salem guardou um N√∫mero Secreto em seu caldeir√£o.")
    print("Voc√™ tem poucas chances para descobrir... ou UM FEITI√áO cair√° sobre voc√™!\n")

    # 2. Configura√ß√£o do jogo
    numero_secreto = random.randint(1, 50)  # n√∫mero aleat√≥rio entre 1 e 50
    tentativas_maximas = 7
    tentativas_restantes = tentativas_maximas
    acertou = False

    # 3. Loop principal do jogo
    while tentativas_restantes > 0 and not acertou:
        print(f"Tentativas restantes: {tentativas_restantes}")

        # Leitura da entrada do jogador
        try:
            chute = int(input("Qual √© o seu palpite entre 1 e 50? "))
        except ValueError:
            print("‚ùå Entrada inv√°lida! Digite apenas n√∫meros inteiros.\n")
            continue  # volta ao in√≠cio sem gastar tentativa

        # 4. Valida√ß√£o do palpite
        if chute < 1 or chute > 50:
            print("‚ö†Ô∏è O palpite deve estar entre 1 e 50. Preste aten√ß√£o!\n")
            continue

        # Gasta uma tentativa v√°lida
        tentativas_restantes -= 1

        # 5. Verifica√ß√£o do palpite
        if chute == numero_secreto:
            print(f"\nüéâ PARAB√âNS! Voc√™ quebrou o feiti√ßo! O n√∫mero era {numero_secreto}!")
            acertou = True
        elif chute > numero_secreto:
            print("üîÆ Muito Alto! O caldeir√£o borbulha dizendo que o n√∫mero √© MENOR...\n")
        else:
            print("üí® Muito Baixo! A fuma√ßa do caldeir√£o indica que o n√∫mero √© MAIOR...\n")

    # 6. Fim do jogo
    if not acertou:
        print("\nüíÄ Suas chances acabaram! O feiti√ßo est√° completo!")
        print(f"O N√∫mero Secreto da Bruxa era {numero_secreto}.")

# 7. Execu√ß√£o principal
if __name__ == "__main__":
    jogar_adivinhacao()
```
