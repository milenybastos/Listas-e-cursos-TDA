```python
#  Lista de Provas Simples com $

asg = {
    1: "Introdução ao Python",
    2: "Estruturas de Decisão",
    3: "Laços de Repetição"
}

while True:
    print("\n Provas Pendentes:")
    if len(asg) == 0:
        print(" Todas as provas foram concluídas!")
        break

    for num, nome in asg.items():
        print(f"{num}. ${nome}")  # Adicionado o $ aqui

    escolha = input("\nDigite o número da prova que concluiu (ou 0 para sair): ")

    if escolha == "0":
        print("Saindo... Bons estudos! ")
        break

    if escolha.isdigit():
        num = int(escolha)
        if num in asg:
            print(f" Prova '{asg[num]}' concluída e removida!")
            del asg[num]
        else:
            print("Número inválido! Tente novamente.")
    else:
        print("Digite um número válido!")
```
